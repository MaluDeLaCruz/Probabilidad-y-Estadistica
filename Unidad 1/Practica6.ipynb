```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
from statistics import *
import matplotlib.pyplot as plt
import math
```

```python
autos = pd.read_csv("autos2.csv", encoding="ISO-8859-1")
```

```python
autos.head()
```

```python
autos.info()
```

```python
#mostrar el precio maximo de la base de datos
autos.price.max()
```

```python
#mostrar el precio minimo de la base de datos
autos.price.min()
```

```python
#mostrar el valor de moda de precios
moda = autos.price.mode()
print(moda)
```

```python
#motrar la media de los precios
media = autos.price.mean()
print(media)
```

```python
#mostrar la mediana de los precios
mediana = autos.price.median()
print(mediana)
```

```python
#determinar el valor de la desviacion estandar de precios
desviacion_estandar_autos = autos.price.std()
print(desviacion_estandar_autos)
```

```python
#mostrar los estadisticos descriptivos y los cuartiles
autos.price.describe()
```

```python
#Realizar un Histograma de precios de los automviles
plt.hist(autos.price)
plt.title("Histograma de Precios de los automviles")
plt.axvline(mediana, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana:.2f}')
plt.axvline(media, color='b', linestyle='dashed', linewidth=2, label=f'Media={media:.2f}')
plt.axvline(desviacion_estandar_autos, color='g', linestyle='dashed', linewidth=2, label=f'Desv. Estandar={desviacion_estandar_autos:.2f}')
plt.xlabel('Precios')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Se puede concluir que el costo mas alto en los automoviles presentados es 45400.0, mientras que el costo ms bajo es de 5118.0

sin embargo el promedio en cuanto a los precio ronda en 13207. Se puede
concluir que esto indica una gran diferencia en los precios , lo que sugiere que hay vehculos de diferentes gamas y caractersticas

```python
#Determinar mediante la regla de sturges la cantidad optima de bins o separadores
n = len(autos)
k = int(1 + math.log2(n))
print("Valor de n = ",n ,"\nNmero de intervalos =",k)
```

```python
#diagrama de caja y bigotes que muestre la relacin
#estadstica entre la columna del precio del automvil (price) y
#tipo de vehculo (body style)
sns.boxplot(x="body-style", y="price", data=autos)
plt.title("Relacin estadstica del precio y el tipo de vehculo")
```

Podemos observar en los autos hatchback, sedan y wagon se presentan datos atpicos en cuanto al precio. Esto sugiere que, dentro de cada categora, existen vehculos cuyo valor se aleja significativamente del rango general de precios

```python
sns.boxplot(x="drive-wheels", y="price", data=autos)
plt.title("Relacin estadstica del precio y el tipo traccin ruedas")
```

Podemos observar como en los 3 tipos de traccion de llantas hay valores atpicos, sin embargo, en 4wd se puede concluir que no hay mucho rango entre el valor minimo y el valor mximo, a diferencia del rwd que si tenemos un rango mayor.

```python
sns.boxplot(x="engine-location", y="price", data=autos)
plt.title("Relacin estadstica del precio y la ubicacin del motor")
```

Podemos concluir que los autos que tienen el motor adelante tienden a tener ms valores atpicos en cuanto el precio, por otro lado, los autos que tienen el motor atrs no tienen mucha diferencia en cuanto el rango de precios.
