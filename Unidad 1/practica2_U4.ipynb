#DISTRIBUCIÓN NORMAL

```python
from scipy.stats import binom
```

```python
probabilidad_exito = 0.83
num_pacientes = 16
x = 16
probabilidad_ninguno_enfermo= binom.pmf( x, num_pacientes, probabilidad_exito)
print("La probabilidad de que ningún paciente sufra la enfermedad es:", probabilidad_ninguno_enfermo)
```

```python
probabilidad_exito = 0.83
num_pacientes = 16
x = 0
probabilidad_todos_enfermos= binom.pmf( x, num_pacientes, probabilidad_exito)
print("La probabilidad de que todos los paciente sufran la enfermedad es:", probabilidad_todos_enfermos)
```

#DISTRIBUCIÓN DE POISSON
##Ejercicios

```python
from scipy.stats import poisson
media_accidentes = 2
probabilidad_0_accidentes = poisson.pmf(0, media_accidentes)
print("1. Probabilidad de que no haya accidentes en 1 semana:", probabilidad_0_accidentes)

media_2_semanas = media_accidentes * 2
probabilidad_3_o_menos_accidentes = poisson.cdf(3, media_2_semanas)
print ("2. Probabilidad de que a lo sumo haya 3 accidentes en 2 semanas:", probabilidad_3_o_menos_accidentes)
```
