```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#generar datos para la distribución normal
media = 70
desv_std = 4
num_datos = 500
#generamos los datos de manera aleatoria y los ponemos en una lista (array)
datos = np.random.normal(media,desv_std, num_datos)
datos[0:50]#mostrar los primeros 50
```

```python
#graficar histograma
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, color='b')
plt.title('Distribución normal')
plt.xlabel('Peso en Kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
#crear un rango de valores para crear una densidad
x = np.linspace(media-3 * desv_std, media+3 * desv_std, num_datos)
#calcular la pdf de la distribucion normal
pdf = 1 / (desv_std * np.sqrt(2 * np.pi)) * np.exp(-(x - media)**2/ (2 * desv_std **2))
```

```python
pdf.size, num_datos
```

```python
print("Imprimir los primeros 50 valores de x",x[0:50]," , y los primeros 50 pdf", pdf[0:50])
```

```python
#visualizar la pdf
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label='Normal:(media:{}; ds:{})'. format(np.round(media, 2), np.round(desv_std)))
plt.title('Función de Densidad de Probabilidad (PDF) - Distribución Normal')
plt.xlabel('Pesos en Kg. de los estudiantes')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot( x, pdf, 'k', linewidth=2)
plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
media=70
desv_std=4
total_estudiantes=500
#variable para calculo de la probabilidad de que un estudiante que pese menos de 75kg
peso_limite=79
prob_menos_75kg=norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_75kg=total_estudiantes * prob_menos_75kg
print('Probabilidad de que un estudiante pese menos de 75kg:',prob_menos_75kg)
print('Numero de estudiantes que pesan menos de 75kg:',estudiantes_menos_75kg)
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std)
y = norm.pdf(x, media, desv_std)
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x<= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
media=70
desv_std=4
total_estudiantes=500
#variable para calculo de la probabilidad de que un estudiante que pese menos de 62kg
peso_limite=62
prob_menos_62kg=norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_62kg=total_estudiantes * prob_menos_62kg
print('Probabilidad de que un estudiante pese menos de 62kg:',prob_menos_62kg)
print('Numero de estudiantes que pesan menos de 62kg:',estudiantes_menos_62kg)
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std)
y = norm.pdf(x, media, desv_std)
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x<= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
media=70
desv_std=4
total_estudiantes=500
#variable para calculo de la probabilidad de que un estudiante que pese mas de 79kg
peso_limite=79
prob_menos_79kg=norm.cdf(peso_limite, media, desv_std)
prob_mas_79kg=1- prob_menos_79kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg
print('Probabilidad de que un estudiante pese mas de 79kg:',prob_mas_79kg)
print('Numero aproximado de estudiantes que pesan mas de 79kg:',estudiantes_mas_79kg)
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)
plt.plot(x,y,'b',linewidth=2)
plt.fill_between(x,y,where=(x>= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 79kg. \n $\mu=70$, $\sigma=4$, x=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

##Se puede concluir que el peso aproximado de los estudiantes está entre los 60 a 75 kilos, sin embargo la probabilidad de que alguno pese mas de 79kg es muy poca, y hay pocos estudiantes arriba de ese peso.
