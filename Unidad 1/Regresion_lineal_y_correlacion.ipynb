```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#Grupo 1

```python
df1 = pd.read_csv("grupo1.csv")
df1.head()
```

```python
#Exploracin inicial de los datos (descripcin estadstica)
df1.describe()
```

```python
plt.hist(df1['Calificacion'])
plt.title("Histrograma de calificaciones")
plt.xlabel("Calificacin")
plt.ylabel("Frecuencia")
plt.show()
```

Modelo de regresin lineal

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
ln = LinearRegression()

#variables independientes o variables predictoras
x = df1[["HorasEstudio"]]

#variables dependientes o variable objetivo a predecir
y = df1["Calificacion"]
x.head(5)
```

```python
ln.fit(x,y)
```

```python
Ypredict = ln.predict(x)
```

```python
#valor de la pendiente
ln.intercept_
```

```python
print("Valor de la pendiente:", ln.coef_)
print("Valor de la intercepcion:", ln.intercept_)

plt.figure(figsize=(6,5))
plt.scatter(x, y, color='blue', label='datos')
plt.plot(x, ln.predict(x), color='red', label='Linea de regresion')
plt.scatter(x, Ypredict, color='green', marker='x', s=100,
            label='Prediccion(8,{0:.2f})'.format(Ypredict[0]))
plt.title('Horas de Estudios vs Calificacin Grupo 1')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificacin')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.set(style="whitegrid")
sns.regplot(x=x, y=y, scatter=True, color='blue')
plt.title('Diagrama de dispersion con linea de regresion')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacin')
plt.grid(True)
plt.show()
```

```python
#coeficiente de correlacin Pearson
from scipy.stats import pearsonr
coeficiente_correlacion, p_valor = pearsonr(df1['HorasEstudio'], df1['Calificacion'])

print("Coeficiente de correlacion de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```

#Preguntas
1. **Explicar cul es la interpretacin del coeficiente de regresin en trminos de la
relacin entre el tiempo de estudio y las calificaciones finales?**
Dado que la lnea de regresin es casi horizontal (pendiente cercana a cero), el coeficiente de regresin es muy bajo, lo que indica que aumentar las horas de estudio no tiene un efecto significativo sobre las calificaciones en este grupo. En otras palabras, el tiempo de estudio no predice bien las calificaciones.
2. **Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y

Visualmente, los puntos estn bastante dispersos y no siguen un patrn lineal claro. Hay una relacin muy dbil (casi nula) entre el tiempo de estudio y la calificacin. No se puede afirmar que una variable aumente consistentemente cuando la otra lo hace.
3. **El modelo de regresin lineal es estadsticamente significativo?**
No, el modelo no es estadsticamente significativo.
4. **Qu tan confiable es el modelo de regresin para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?**
Baja confiabilidad.
Dado que los datos estn muy dispersos y la lnea de regresin es casi horizontal
