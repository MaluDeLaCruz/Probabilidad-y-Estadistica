```python
from statistics import *
```

```python
#Lista de calificaciones del grupo a de probabilidad y estadistica
grupoa=[70,70,65,90,100,95,45,80,73,20,100,96]
print(grupoa)
```

```python
#Sacar la media de las calificaciones con una formula
promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
#utilizar la funcion mean que esta en la libreria statistics para sacar el promedio
promedio2 = mean(grupoa)
print(promedio2)
```

```python
#Determinar el valor de la mediana con la funcion median
mediana = median(grupoa)
print(mediana)
```

```python
#Determinar la moda con la funcion mode
moda = mode(grupoa)
print(moda)
```

```python
#determinar multiples modas en caso de que existan con la funcion multimode
moda2 = multimode(grupoa)
print(moda2)
```

#EJEMPLO 2. TIEMPOS DE 100 METROS DE UN GRUPO DE 12 ALUMNOS

```python
tiempos100mts = [15.10,17.20,14.69,13.27,22.15,18.71,19.15,20.65,15.10,17.20]
print(tiempos100mts)
```

```python
#calcular el promedio
print("El tiempo promedio de los corredores es =>", mean(tiempos100mts))
```

```python
#calcular la mediana
mediana = median(tiempos100mts)
print("La media de los tiempos de los corredores es =>", mediana)
```

```python
#calcular la moda
moda = mode(tiempos100mts)
print("La moda de los tiempos de los corredores es =>", moda)
```

```python
#calcular multimodas
multimoda = multimode(tiempos100mts)
print("La moda de los tiempos de los corredores es =>", multimoda)
```

#EJEMPLO 3. Histogramas de 2 grupos

```python
#libreria para generar numeros aleatorios
import random
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50,100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50,100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
#importat libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la Fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la Fabrica 2")
```
