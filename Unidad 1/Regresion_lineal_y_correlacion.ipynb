```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#Grupo 1

```python
df1 = pd.read_csv("grupo1.csv")
df1.head()
```

```python
#Exploración inicial de los datos (descripción estadística)
df1.describe()
```

```python
plt.hist(df1['Calificacion'])
plt.title("Histrograma de calificaciones")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.show()
```

Modelo de regresión lineal

```python
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
ln = LinearRegression()

#variables independientes o variables predictoras
x = df1[["HorasEstudio"]]

#variables dependientes o variable objetivo a predecir
y = df1["Calificacion"]
x.head(5)
```

```python
ln.fit(x,y)
```

```python
Ypredict = ln.predict(x)
```

```python
#valor de la pendiente
ln.intercept_
```

```python
print("Valor de la pendiente:", ln.coef_)
print("Valor de la intercepcion:", ln.intercept_)

plt.figure(figsize=(6,5))
plt.scatter(x, y, color='blue', label='datos')
plt.plot(x, ln.predict(x), color='red', label='Linea de regresion')
plt.scatter(x, Ypredict, color='green', marker='x', s=100,
            label='Prediccion(8,{0:.2f})'.format(Ypredict[0]))
plt.title('Horas de Estudios vs Calificación Grupo 1')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.set(style="whitegrid")
sns.regplot(x=x, y=y, scatter=True, color='blue')
plt.title('Diagrama de dispersion con linea de regresion')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()
```

```python
#coeficiente de correlación Pearson
from scipy.stats import pearsonr
coeficiente_correlacion, p_valor = pearsonr(df1['HorasEstudio'], df1['Calificacion'])

print("Coeficiente de correlacion de Pearson:", coeficiente_correlacion)
print("Valor p:", p_valor)
```

#Preguntas
1. **¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la
relación entre el tiempo de estudio y las calificaciones finales?**
Dado que la línea de regresión es casi horizontal (pendiente cercana a cero), el coeficiente de regresión es muy bajo, lo que indica que aumentar las horas de estudio no tiene un efecto significativo sobre las calificaciones en este grupo. En otras palabras, el tiempo de estudio no predice bien las calificaciones.
2. **¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y
dirección de la relación entre las variables?**
Visualmente, los puntos están bastante dispersos y no siguen un patrón lineal claro. Hay una relación muy débil (casi nula) entre el tiempo de estudio y la calificación. No se puede afirmar que una variable aumente consistentemente cuando la otra lo hace.
3. **¿El modelo de regresión lineal es estadísticamente significativo?**
No, el modelo no es estadísticamente significativo.
4. **¿Qué tan confiable es el modelo de regresión para predecir las
calificaciones finales de los estudiantes basado en el tiempo de estudio?**
Baja confiabilidad.
Dado que los datos están muy dispersos y la línea de regresión es casi horizontal

#CONCLUSIONES GENERALES
A partir del análisis de los tres grupos, se observa que la relación entre las horas de estudio y las calificaciones varía considerablemente.
